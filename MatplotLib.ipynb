- 'd': diamond
- 'D': Diamond
- 'p': pentagon

- 'H': hexagon
- 'v': triangle down

- '>': triangel right

## Line

- '-': solid line
- ':': dotted line
- '--': dashed line
- '-.': dash/dott line
- ' ': None


````python
plt.plot(x,y,'o:')
plt.show()
````

````python
plt.plot(x,y,'o--')
plt.show()
````

## Colors

- 'r': red

## Marker Size

paramater ```ms``` which is short for Marker Size



````python
plt.plot(x,y,'o:k', ms = 25)
plt.show()
````

## Marker Color
- mec: marker edge color
- mfc: marker face color


[Other available colors](https://www.w3schools.com/colors/colors_names.asp)

````python
plt.plot(x,y,'|', ms = 25, mec = 'g', mfc = 'w')
plt.show()
````

## Linestyle

- linestyle/ls: dotted, dashed, dotted/dashed, solid. 



````python
plt.plot(x,y, 'p', ls = ' ',  ms = 25, mec = 'g', mfc = 'w')
plt.show()
````

## Line Color

- color/c: and mention the color in single quotes


[Other available colors](https://www.w3schools.com/colors/colors_names.asp)


## plot with marker as a circle in red color and line style solid and red in color


````python
plt.plot(x,y, marker = 'o', ls = '-', ms = 20, mec = 'k', mfc = 'Gainsboro', c = 'r')
plt.show()
````

## Line width

- linewidth/lw

````python
plt.plot(x,y, marker = 'o', ls = '-', ms = 20, mec = 'k', mfc = 'Gainsboro', c = 'r', lw = 35)
plt.show()
````

##Multiple plots in the same graph

````python
x1 = [2,3,6,5,9,8,4,6,2,3,1,5]
y1 = [5,6,3,2,6,9,8,7,4,5,2,1]

plt.plot(x, y, 'o:')
plt.plot(x1, y1, 'o:')
plt.show()
````

## Change Scale of a plot


````python
x1 = [2,3,6,5,9,8,4,6,2,3,1,5]
y1 = [5,6,3,2,6,9,8,7,4,5,2,1]

plt.plot(x, y, 'o:')
plt.plot(x1, y1, 'o:')

plt.xlim(0, 15)
plt.ylim(0, 20)

plt.show()
````

## Labels for plots

````python
x1 = [2,3,6,5,9,8,4,6,2,3,1,5]
y1 = [5,6,3,2,6,9,8,7,4,5,2,1]

plt.plot(x, y, 'o:')
plt.plot(x1, y1, 'o:')

plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")

plt.show()
````

## Title to the plot

````python
x1 = [2,3,6,5,9,8,4,6,2,3,1,5]
y1 = [5,6,3,2,6,9,8,7,4,5,2,1]

plt.plot(x, y, 'o:')
plt.plot(x1, y1, 'o:')

plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")

plt.title("This is a plot")

plt.show()
````

## Position the title

- loc parameter is used


````python
x1 = [2,3,6,5,9,8,4,6,2,3,1,5]
y1 = [5,6,3,2,6,9,8,7,4,5,2,1]

plt.plot(x, y, 'o:')
plt.plot(x1, y1, 'o:')

plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")

plt.title("This is a plot", loc = 'right')

plt.show()
````

## Change the font properties of the title

fontdict is the parameter used to change the fort properties

````python
x1 = [2,3,6,5,9,8,4,6,2,3,1,5]
y1 = [5,6,3,2,6,9,8,7,4,5,2,1]

plt.plot(x, y, 'o:')
plt.plot(x1, y1, 'o:')

plt.xlabel("This is X axis", fontdict={'color':'red'})
plt.ylabel("This is Y axis", fontdict={'color':'blue'})

plt.title("This is a plot", loc = 'left', fontdict={'color':'brown', 'size':25})

plt.show()
````
