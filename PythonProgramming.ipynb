#Fizz Buzz
given a value n
- print every number from 1 to n
- if the number is divisible by 3, **print Fizz**
- if the number is divisible by 5, **print Buzz**
- if the number is divisible by 3 and 5, print **Fizz Buzz**

```
Algorithm:
- we check if the number is divisible by both 3 and 5
- if yes we print fizz buzz and then continue with the next number
- if not we compare with 5
- if yes we print buzz and continue to the nubmer
- if not we compare with 3
- if yes we print fizz and continue to the next tumber
- if not print the numbre
```

````python
n = int(input())
i = 1
while i<=n:
  if i%3 == 0 and i%5 == 0:
    print("Fizz Buzz")
  elif i%3 == 0:
    print("Fizz")
  elif i%5 == 0:
    print("Buzz")
  else:
    print(i)
  i = i+1
````

#Calculator Program

##1. Simple calculator program

````python
a = int(input("Enter a: "))
op = input("Enter the operation: ")
b = int(input("Enter b: "))

if op == '+':
  print(a, op, b, '=', a+b)
elif op == '-':
  print(a, op, b, '=', a-b)
elif op == '*':
  print(a, op, b, '=', a*b)
elif op == '/':
  print(a, op, b, '=', a/b)
else:
  print("Enter the rigth operator")
````
